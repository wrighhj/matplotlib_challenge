```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
from matplotlib.pylab import rcParams
import seaborn as sns
sns.set_style('ticks')
%config InLineBackend.figure_formats = ['retina']
plt.rc('lines', linewidth = 1)
rcParams['figure.figsize'] = 10,8

# Study data files
# Read the mouse data and the study results
mice = pd.read_csv("Resources/Mouse_metadata.csv")
results = pd.read_csv("Resources/Study_results.csv")

# Combine the data into a single dataset
scc = pd.merge(mice, results, on="Mouse ID", how="outer")
# Display the data table for preview
scc.head(100)
```

```python
# Checking the number of mice.
len(scc['Mouse ID'].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
scc_1 = scc.loc[scc.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
scc_1
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_scc = scc[scc['Mouse ID'].isin(scc_1)==False] 
clean_scc
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
tumors = clean_scc[clean_scc['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin', 'Placebo'])].sort_values(['Timepoint'], ascending = False).rename(columns= {'Tumor Volume (mm3)': 'Final Tumor Volume (mm3)'}).reset_index()
tumors.head(63)
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_scc = pd.merge(clean_scc, tumors, how = "outer").dropna()
final_scc
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = pd.Series(final_scc['Drug Regimen']).unique().tolist()
drugs
```

```python
# Create empty list to fill with tumor vol data (for plotting)
# Heather note: Including Placebo to see the control with everything else
# Heather note: these lists are just FINAL tumor volume data

Ramicane_tumors = []
Ramicane_mice = []
for i in final_scc.index:
    if (final_scc['Drug Regimen'][i] == "Ramicane"):
        Ramicane_tumors.append(final_scc['Final Tumor Volume (mm3)'][i])
        Ramicane_mice.append(final_scc['Mouse ID'][i])
        
Capomulin_tumors = []
Capomulin_mice = []
for i in final_scc.index:
    if (final_scc['Drug Regimen'][i] == "Capomulin"):
        Capomulin_tumors.append(final_scc['Final Tumor Volume (mm3)'][i])
        Capomulin_mice.append(final_scc['Mouse ID'][i])
        
Infubinol_tumors = []
Infubinol_mice = []
for i in final_scc.index:
    if (final_scc['Drug Regimen'][i] == "Infubinol"):
        Infubinol_tumors.append(final_scc['Final Tumor Volume (mm3)'][i])
        Infubinol_mice.append(final_scc['Mouse ID'][i])
        
Placebo_tumors = []
Placebo_mice = []
for i in final_scc.index:
    if (final_scc['Drug Regimen'][i] == "Placebo"):
        Placebo_tumors.append(final_scc['Final Tumor Volume (mm3)'][i])
        Placebo_mice.append(final_scc['Mouse ID'][i])
        
Ceftamin_tumors = []
Ceftamin_mice = []
for i in final_scc.index:
    if (final_scc['Drug Regimen'][i] == "Ceftamin"):
        Ceftamin_tumors.append(final_scc['Final Tumor Volume (mm3)'][i])
        Ceftamin_mice.append(final_scc['Mouse ID'][i])
        
# FTV = Final Tumor Volume
# Making dataframes for boxplots below
ramicane_df = pd.DataFrame(Ramicane_tumors, Ramicane_mice, columns=['FTV']).reset_index().rename(columns= {"index": "Mouse ID"})
ramicane_df

capomulin_df = pd.DataFrame(Capomulin_tumors, Capomulin_mice, columns=['FTV']).reset_index().rename(columns= {"index": "Mouse ID"})
capomulin_df

infubinol_df = pd.DataFrame(Infubinol_tumors, Infubinol_mice, columns=['FTV']).reset_index().rename(columns= {"index": "Mouse ID"})
infubinol_df

placebo_df = pd.DataFrame(Placebo_tumors, Placebo_mice, columns=['FTV']).reset_index().rename(columns= {"index": "Mouse ID"})
placebo_df

ceftamin_df = pd.DataFrame(Ceftamin_tumors, Ceftamin_mice, columns=['FTV']).reset_index().rename(columns= {"index": "Mouse ID"})
ceftamin_df
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Heather note: Including Placebo to see the control with everything else

# calculating Placebo outliers:
Placebo_tvq = placebo_df['FTV'].quantile([.25, .5, .75])
lowerq = Placebo_tvq[.25]
upperq = Placebo_tvq[.75]
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Possible Placebo outliers below: {lower_bound}")
print(f"Possible Placebo outliers above: {upper_bound}")

# calculating Ramicane outliers:
Ramicane_tvq = ramicane_df['FTV'].quantile([.25, .5, .75])
lowerq = Ramicane_tvq[.25]
upperq = Ramicane_tvq[.75]
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Possible Ramicane outliers below: {lower_bound}")
print(f"Possible Ramicane outliers above: {upper_bound}")

# calculating Capomulin outliers:
Capomulin_tvq = capomulin_df['FTV'].quantile([.25, .5, .75])
lowerq = Capomulin_tvq[.25]
upperq = Capomulin_tvq[.75]
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Possible Capomulin outliers below: {lower_bound}")
print(f"Possible Capomulin outliers above: {upper_bound}")

# calculating Infubinol outliers:
Infubinol_tvq = infubinol_df['FTV'].quantile([.25, .5, .75])
lowerq = Infubinol_tvq[.25]
upperq = Infubinol_tvq[.75]
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Possible Infubinol outliers below: {lower_bound}")
print(f"Possible Infubinol outliers above: {upper_bound}")

# calculating Ceftamin outliers:
Ceftamin_tvq = ceftamin_df['FTV'].quantile([.25, .5, .75])
lowerq = Ceftamin_tvq[.25]
upperq = Ceftamin_tvq[.75]
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Possible Ceftamin outliers below: {lower_bound}")
print(f"Possible Ceftamin outliers above: {upper_bound}")
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

# Calculate average tumor volume for each drug timepoint
tumor_volume = pd.DataFrame(tumors.groupby(['Drug Regimen', 'Timepoint'])['Final Tumor Volume (mm3)'].mean()).reset_index()
tumor_volume

# Make a pivot table so all the averages are nice to look at
pivot_tumor_volume = tumor_volume.pivot(index = "Timepoint", columns = "Drug Regimen", values = "Final Tumor Volume (mm3)")
ptv = pivot_tumor_volume[['Placebo', 'Capomulin', 'Ceftamin', 'Infubinol', 'Ramicane']]
ptv

# plot Placebo vs Capomulin 
plt.errorbar(ptv.index,ptv['Placebo'])
plt.errorbar(ptv.index,ptv['Capomulin'], color = "red")
plt.title("Growth Curves")
plt.ylabel('Average Tumor Volume (mm3)')
plt.xlabel('Time (Days)')
plt.legend(loc = "best", labels = ('Placebo', 'Capomulin'))
plt.savefig("Images/growth.png")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
mouse_weight = pd.DataFrame(tumors.groupby(['Drug Regimen', 'Timepoint'])['Weight (g)'].mean()).reset_index()
mouse_weight
mouse_tumor = pd.DataFrame(tumors.groupby(['Drug Regimen', 'Timepoint'])['Final Tumor Volume (mm3)'].mean()).reset_index()
mouse_tumor
tm_df = pd.merge(mouse_weight, mouse_tumor, how = "outer", on = ['Drug Regimen', 'Timepoint'])
# tm_df1 = tm_df[tm_df['Drug Regimen'].isin(['Placebo', 'Capomulin'])]
tm_df1 = tm_df[tm_df['Drug Regimen'].isin(['Capomulin'])]
tm_df1

plt.scatter(tm_df1['Final Tumor Volume (mm3)'], tm_df1['Weight (g)'],c=tm_df1['Timepoint'])
clb = plt.colorbar()
clb.set_label("Time (days)")
plt.title("Weight vs Tumor Volume in Capomulin Treatment Arm")
plt.ylabel('Weight (g)')
plt.xlabel('Tumor Volume (mm3)')
plt.savefig("Images/box_whisk.png")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
tumor = tm_df1.iloc[:,3]
weight = tm_df1.iloc[:,2]
correlation = round(st.pearsonr(tumor,weight)[0],2)
print(f"The correlation coefficient between tumor volume and mouse weight in the Capomulin treatment arm is {correlation}")
# Calculate the linear regression model 
tm_df1
x = tm_df1['Final Tumor Volume (mm3)']
y = tm_df1['Weight (g)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress_values = tm_df1['Final Tumor Volume (mm3)'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
r_square = rvalue**2
# combine graph and linear regression model
plt.scatter(x,y,c=tm_df1['Timepoint'])
clb = plt.colorbar()
clb.set_label("Time (days)")
plt.plot(x,regress_values,"black")
plt.annotate(line_eq,(42,19.9),fontsize = 15, color="black")
plt.annotate(f"r^2= {r_square}",(41,19.85),fontsize = 15, color="black")
clb.set_label("Time (days)")
plt.title("Weight vs Tumor Volume in Capomulin Treatment Arm")
plt.ylabel('Weight (g)')
plt.xlabel('Tumor Volume (mm3)')
plt.savefig("Images/linregress.png")
```
